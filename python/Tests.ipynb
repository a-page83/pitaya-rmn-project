```python
%matplotlib qt
import struct
import os
import matplotlib.pyplot as plt 
fileName = 'test3.bin'
with open(fileName, mode='rb') as file: # b is important -> binary
    fileContent = file.read()





headerbin = fileContent[0:16]
header = struct.unpack("iiii",headerbin)
print(header)
size = header[0]

values = struct.unpack("<"+"h"*size,fileContent[16:])
print(values)
channel_A = values[::2]
channel_B = values[1::2]
plt.plot(channel_B, marker='+', linestyle='-', linewidth=1,label='B')

plt.plot(channel_A, marker='+', linestyle='-', linewidth=1,label='A')
plt.legend()
plt.show()
```
